```python
import json
import pandas as pd
```

```python
with open("D:\\ZeynepStudy\\Images\\prompts2numbered.json", 'r') as f:
        prompts = json.load(f)
```

```python
prompts = prompts['prompts']
```

```python
prompts[0]
```

```python
categories = []
indices = []
promptArray = []

for i in range(240):
    categories.append(prompts[i]['category'])
    indices.append(i+1)
    promptArray.append(prompts[i]['prompt'])
```

```python
dict = {'index': indices, 'categories': categories, 'prompts': promptArray}

array = pd.DataFrame(dict)
```

```python
array.to_excel("D:\\ZeynepStudy\\Images\\paintingPrompts.xlsx")
```
