```python
conditions = {
    0: "alone",
    1: "with other participant",
    2: "with computer avatar",
    3: "with computer avatar" 
}
```

```python
MemResultsAll = pd.DataFrame()

for participant in participants:    
    # Open the relevant files for the participant
    MemResultsFile = glob.glob("/Users/vojtasmekal/Documents/MemTask/p" + participant + "/*.csv")
    imgCondOrder = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/expConditions.json"

    
    if (participant == '03') | (participant == '04') | (participant == '05'):
        soundOrder1 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_1.json"
        soundOrder2 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_2.json"
        soundOrder3 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_3.json"
    else:
        soundOrder1 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_0.json"
        soundOrder2 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_1.json"
        soundOrder3 = "/Users/vojtasmekal/Documents/MemTask/p" + participant + "/sound_schedule_balanced_with_triggers_2.json"
    
    # Create dataframes for each block's paintings and populate them with paintings, their types, and their timing
    f = open(imgCondOrder)
    imgCondOrderData = json.load(f)
    imageOrder = imgCondOrderData['images_order']
    condOrder = imgCondOrderData['conditions_order']
    imageTimes = range(0,196,5)
    imageDf1 = pd.DataFrame(imageOrder[0:40], columns=['images'])
    imageDf2 = pd.DataFrame(imageOrder[40:80], columns=['images'])
    imageDf3 = pd.DataFrame(imageOrder[80:120], columns=['images'])
    
    imageDf1.insert(0, "time", imageTimes, True)
    imageDf1.insert(2, "axis", 1, True)
    imageDf1["imageCat"] = imageDf1["images"]
    imageDf1.loc[imageDf1["images"].isin(woodscapes), "imageCat"] = 0
    imageDf1.loc[imageDf1["images"].isin(seascapes), "imageCat"] = 1
    imageDf1.loc[imageDf1["images"].isin(single_person), "imageCat"] = 2
    imageDf1.loc[imageDf1["images"].isin(multiple_people), "imageCat"] = 3
    imageDf2.insert(0, "time", imageTimes, True)
    imageDf2.insert(2, "axis", 1, True)
    imageDf2["imageCat"] = imageDf2["images"]
    imageDf2.loc[imageDf2["images"].isin(woodscapes), "imageCat"] = 0
    imageDf2.loc[imageDf2["images"].isin(seascapes), "imageCat"] = 1
    imageDf2.loc[imageDf2["images"].isin(single_person), "imageCat"] = 2
    imageDf2.loc[imageDf2["images"].isin(multiple_people), "imageCat"] = 3
    imageDf3.insert(0, "time", imageTimes, True)
    imageDf3.insert(2, "axis", 1, True)
    imageDf3["imageCat"] = imageDf3["images"]
    imageDf3.loc[imageDf3["images"].isin(woodscapes), "imageCat"] = 0
    imageDf3.loc[imageDf3["images"].isin(seascapes), "imageCat"] = 1
    imageDf3.loc[imageDf3["images"].isin(single_person), "imageCat"] = 2
    imageDf3.loc[imageDf3["images"].isin(multiple_people), "imageCat"] = 3
    
    
    # Create dataframes for each block's sounds and populate them with the sounds, their types, and timing
    f = open(soundOrder1)
    soundsBlock1Data = json.load(f)
    soundTimes = []
    soundList = []
    
    for i in range(len(soundsBlock1Data)):
        soundTimes.append(soundsBlock1Data[i]['time_from_start'])
        soundList.append(soundsBlock1Data[i]['file'])
        
    soundList = [sound[6:-5] for sound in soundList]
    soundDf1 = pd.DataFrame(soundTimes, columns=['time'])
    soundDf1.insert(1, "sound", soundList, True)
    soundDf1.insert(2, "axis", 2, True)
    soundDf1["soundType"] = soundDf1["sound"]
    soundDf1.loc[soundDf1["sound"].isin(stressfulSounds), "soundType"] = 4
    soundDf1.loc[soundDf1["sound"].isin(peacefulSounds), "soundType"] = 5
    
    
    f = open(soundOrder2)
    soundsBlock2Data = json.load(f)
    soundTimes = []
    soundList = []
    
    
    for i in range(len(soundsBlock2Data)):
        soundTimes.append(soundsBlock2Data[i]['time_from_start'])
        soundList.append(soundsBlock2Data[i]['file'])
        
    soundList = [sound[6:-5] for sound in soundList]
    soundDf2 = pd.DataFrame(soundTimes, columns=['time'])
    soundDf2.insert(1, "sound", soundList, True)
    soundDf2.insert(2, "axis", 2, True)
    soundDf2["soundType"] = soundDf2["sound"]
    soundDf2.loc[soundDf2["sound"].isin(stressfulSounds), "soundType"] = 4
    soundDf2.loc[soundDf2["sound"].isin(peacefulSounds), "soundType"] = 5
    
    
    f = open(soundOrder3)
    soundsBlock3Data = json.load(f)
    soundTimes = []
    soundList = []
    
    for i in range(len(soundsBlock3Data)):
        soundTimes.append(soundsBlock3Data[i]['time_from_start'])
        soundList.append(soundsBlock3Data[i]['file'])
        
    soundList = [sound[6:-5] for sound in soundList]
    soundDf3 = pd.DataFrame(soundTimes, columns=['time'])
    soundDf3.insert(1, "sound", soundList, True)
    soundDf3.insert(2, "axis", 2, True)
    soundDf3["soundType"] = soundDf3["sound"]
    soundDf3.loc[soundDf3["sound"].isin(stressfulSounds), "soundType"] = 4
    soundDf3.loc[soundDf3["sound"].isin(peacefulSounds), "soundType"] = 5
    
    
    # Identify the social presence condition for each block
    condBlock1 = conditions[condOrder[0]]
    condBlock2 = conditions[condOrder[1]]
    condBlock3 = conditions[condOrder[2]]
    
    
    # Plot the sounds, paintings, and conditions per block
    fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=False)
    fig.suptitle('Experimental Conditions p' + participant)
    fig.tight_layout(pad=0.7)
    fig.set_figwidth(17)
    fig.set_figheight(8)
    
    ax1.plot('time', 'axis', 's', c='#228833',data=imageDf1.loc[imageDf1['imageCat']==0])
    ax1.plot('time', 'axis', 's', c='#66CCEE',data=imageDf1.loc[imageDf1['imageCat']==1])
    ax1.plot('time', 'axis', 's', c='#AA3377',data=imageDf1.loc[imageDf1['imageCat']==2])
    ax1.plot('time', 'axis', 's', c='#CCBB44',data=imageDf1.loc[imageDf1['imageCat']==3])
    ax1.plot('time', 'axis', 'o', c='#EE6677', data=soundDf1.loc[soundDf1['soundType']==4])
    ax1.plot('time', 'axis', 'o', c='#4477AA', data=soundDf1.loc[soundDf1['soundType']==5])
    ax1.set_ylim([0,3])
    ax1.axes.get_yaxis().set_visible(False)
    ax1.text(-8,1.2,'images')
    ax1.text(-8,2.2,'sounds')
    ax1.set_title('Block 1, condition: ' + condBlock1)
    ax1.grid(visible=True, which='both', axis='x')
    
    ax2.plot('time', 'axis', 's', c='#228833',data=imageDf2.loc[imageDf2['imageCat']==0])
    ax2.plot('time', 'axis', 's', c='#66CCEE',data=imageDf2.loc[imageDf2['imageCat']==1])
    ax2.plot('time', 'axis', 's', c='#AA3377',data=imageDf2.loc[imageDf2['imageCat']==2])
    ax2.plot('time', 'axis', 's', c='#CCBB44',data=imageDf2.loc[imageDf2['imageCat']==3])
    ax2.plot('time', 'axis', 'o', c='#EE6677', data=soundDf2.loc[soundDf2['soundType']==4])
    ax2.plot('time', 'axis', 'o', c='#4477AA', data=soundDf2.loc[soundDf2['soundType']==5])
    ax2.set_ylim([0,3])
    ax2.axes.get_yaxis().set_visible(False)
    ax2.text(-8,1.2,'images')
    ax2.text(-8,2.2,'sounds')
    ax2.set_title('Block 2, condition: ' + condBlock2)
    ax2.grid(visible=True, which='both', axis='x')
    
    line1, = ax3.plot('time', 'axis', 's', c='#228833', label='Woodscapes',data=imageDf3.loc[imageDf3['imageCat']==0])
    line2, = ax3.plot('time', 'axis', 's', c='#66CCEE', label='Seascapes',data=imageDf3.loc[imageDf3['imageCat']==1])
    line3, = ax3.plot('time', 'axis', 's', c='#AA3377', label='Single Person',data=imageDf3.loc[imageDf3['imageCat']==2])
    line4, = ax3.plot('time', 'axis', 's', c='#CCBB44', label='Multiple People',data=imageDf3.loc[imageDf3['imageCat']==3])
    line5, = ax3.plot('time', 'axis', 'o', c='#EE6677', label='Threatening', data=soundDf3.loc[soundDf3['soundType']==4])
    line6, = ax3.plot('time', 'axis', 'o', c='#4477AA', label='Peaceful', data=soundDf3.loc[soundDf3['soundType']==5])
    ax3.set_ylim([0,3])
    ax3.axes.get_yaxis().set_visible(False)
    ax3.text(-8,1.2,'images')
    ax3.text(-8,2.2,'sounds')
    ax3.set_title('Block 3, condition: ' + condBlock3)
    plt.xlabel('Time (sec)')
    plt.grid(visible=True, which='both', axis='x')
    fig.legend(handles=[line1, line2, line3, line4, line5, line6], bbox_to_anchor=(0.06, 0.59))
    
    fig.savefig('/Users/vojtasmekal/Documents/output/MemTask/'+participant+'_expConditions.png', bbox_inches='tight')
    
    
    # Prepare dataframe for memory task results
    paintings = list(range(1,241))
    paintingMemory = pd.DataFrame(paintings, columns=['painting'])
    
    paintingMemory['paintingCat'] = paintingMemory['painting']
    paintingMemory['paintingOld'] = 0
    paintingMemory.loc[paintingMemory['painting'].isin(woodscapes), 'paintingCat'] = 'woodscape'
    paintingMemory.loc[paintingMemory['painting'].isin(seascapes), 'paintingCat'] = 'seascape'
    paintingMemory.loc[paintingMemory['painting'].isin(single_person), 'paintingCat'] = 'single_person'
    paintingMemory.loc[paintingMemory['painting'].isin(multiple_people), 'paintingCat'] = 'multiple_people'
    paintingMemory.loc[paintingMemory['painting'].isin(oldPaintings), 'paintingOld'] = 1
    paintingMemory['paintingOrder'] = 0
    
    for i in range(len(oldPaintings)):
        paintingMemory.loc[paintingMemory['painting']==oldPaintings[i], 'paintingOrder'] = imageOrder.index(oldPaintings[i]) +1
    
    paintingMemory['sound'] = 0
    paintingMemory['expCond'] = 0
    paintingMemory['participant'] = participant
    
    
    # Check to see if a sound was played while a painting was being shown and record which sound if yes
    for i in range(len(soundDf1)):
        tempTime = soundDf1['time'][i]
        paintingTime = (tempTime//5)*5
        tempPainting = imageDf1['images'].loc[imageDf1['time']==paintingTime].reset_index()
        tempPainting = tempPainting['images'].to_list()
        paintingMemory.loc[paintingMemory['painting']==tempPainting[0], 'sound'] = soundDf1['sound'][i]
        
    for i in range(len(soundDf2)):
        tempTime = soundDf2['time'][i]
        paintingTime = (tempTime//5)*5
        tempPainting = imageDf2['images'].loc[imageDf2['time']==paintingTime].reset_index()
        tempPainting = tempPainting['images'].to_list()
        paintingMemory.loc[paintingMemory['painting']==tempPainting[0], 'sound'] = soundDf2['sound'][i]
          
    for i in range(len(soundDf3)):
        tempTime = soundDf3['time'][i]
        paintingTime = (tempTime//5)*5
        tempPainting = imageDf3['images'].loc[imageDf3['time']==paintingTime].reset_index()
        tempPainting = tempPainting['images'].to_list()
        paintingMemory.loc[paintingMemory['painting']==tempPainting[0], 'sound'] = soundDf3['sound'][i]
        
        
    # Identify which social presence condition each painting was presented in
    for i in range(len(imageDf1)):
        tempImage = imageDf1['images'][i]
        paintingMemory.loc[paintingMemory['painting']==tempImage, 'expCond'] = conditions[condOrder[0]]
    for i in range(len(imageDf2)):
        tempImage = imageDf2['images'][i]
        paintingMemory.loc[paintingMemory['painting']==tempImage, 'expCond'] = conditions[condOrder[1]]
    for i in range(len(imageDf3)):
        tempImage = imageDf3['images'][i]
        paintingMemory.loc[paintingMemory['painting']==tempImage, 'expCond'] = conditions[condOrder[2]]
        
    
    # Identify what type of sound was played with each painting
    paintingMemory['soundType'] = 0
    paintingMemory.loc[paintingMemory["sound"].isin(stressfulSounds), "soundType"] = 'stressful'
    paintingMemory.loc[paintingMemory["sound"].isin(peacefulSounds), "soundType"] = 'peaceful'
    paintingMemory.loc[paintingMemory["sound"]==0, "soundType"] = 'no sound'
    
    # Identify which artist inspired each painting
    paintingMemory['artist'] = 0
    paintingMemory.loc[paintingMemory["painting"].isin(renoir), 'artist'] = 'Renoir'
    paintingMemory.loc[paintingMemory["painting"].isin(picasso), 'artist'] = 'Picasso'
    paintingMemory.loc[paintingMemory["painting"].isin(toulouseLautrec), 'artist'] = 'Toulouse-Lautrec'
    paintingMemory.loc[paintingMemory["painting"].isin(chagall), 'artist'] = 'Chagall'
    paintingMemory.loc[paintingMemory["painting"].isin(vanGogh), 'artist'] = 'van Gogh'
    paintingMemory.loc[paintingMemory["painting"].isin(turner), 'artist'] = 'Turner'
    paintingMemory.loc[paintingMemory["painting"].isin(vermeer), 'artist'] = 'Vermeer'
    paintingMemory.loc[paintingMemory["painting"].isin(manet), 'artist'] = 'Manet'
    paintingMemory.loc[paintingMemory["painting"].isin(raphael), 'artist'] = 'Raphael'
    paintingMemory.loc[paintingMemory["painting"].isin(kahlo), 'artist'] = 'Kahlo'
    paintingMemory.loc[paintingMemory["painting"].isin(klimt), 'artist'] = 'Klimt'
    
    
    # Populate the memory results dataframe with the responses, reaction times, confidence levels, and order for each painting
    MemResults = pd.read_csv(MemResultsFile[0])
    
    MemResults = MemResults.iloc[1:]
    MemResults['trials.thisIndex'] = MemResults['trials.thisIndex'] + 1
    
    paintingMemory['response'] = 0
    paintingMemory['reactionTime'] = 0
    paintingMemory['confidence'] = 0
    
    for i in range(len(MemResults)):
        trialPainting = MemResults['trials.thisIndex'][i+1]
        paintingMemory.loc[paintingMemory['painting']==trialPainting, 'response'] = MemResults['key_resp.keys'][i+1]
        paintingMemory.loc[paintingMemory['painting']==trialPainting, 'reactionTime'] = MemResults['memoryTrial.stopped'][i+1] - MemResults['memoryTrial.started'][i+1]
        paintingMemory.loc[paintingMemory['painting']==trialPainting, 'confidence'] = MemResults['slider.response'][i+1]
        
    paintingMemory['correct'] = 0
    paintingMemory.loc[(paintingMemory['response']==1) & (paintingMemory['painting'].isin(oldPaintings)), 'correct'] = 1
    paintingMemory.loc[(paintingMemory['response']==2) & (~paintingMemory['painting'].isin(oldPaintings)), 'correct'] = 1
    
    MemResultsAll = pd.concat([MemResultsAll, paintingMemory], ignore_index=True)
```
